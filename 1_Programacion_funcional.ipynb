```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import Pablo as pb
```

```python
#Cargar el archivo .csv

df = pb.cargar_dataset('Sydney.csv')
df
```

```python
df.info()
```

```python
#Eliminar columnas vacías 
df.drop(['neighbourhood_group_cleansed','calendar_updated'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
valores_nulos = pb.valores_nulos(df)
valores_nulos
```

```python
df["price"] =  df["price"].str.replace(r"[$,]","",regex=True).astype(float)
```

```python
#Con el promedio podemos rellenar los datos nulos y no generar un sesgo ya que toma en cuenta todos los valores de la columna
pb.sustitucion_mean(df,['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value'])
```

```python
# La mediana nos ayuda a mantener los números de las columnas entre los valores altos y bajos para poder tener números lógicos en las columnas
pb.sustitucion_median(df, ['price', 'host_listings_count'])
```

```python
# Con una constante podemos rellenar los datos con un mismo número 
pb.sustitucion_constante(df, ['reviews_per_month', 'host_total_listings_count'])
```

```python
# Este metodo ayuda a que sea una sustitución constante y que los datos tengan cierto orden y sentido
pb.sustitucion_ffill(df,['bathrooms', 'bedrooms', 'beds', 'host_since'])
```

```python
# Con el método de sustitución con una string constante podemos negar los datos nulos, en este caso los booleanos (Falso y Verdadero) podemos rellenarlos de una manera concreta
pb.sustitucion_string(df, ['has_availability'])
```

```python
#Este método evita que los datos se distorcionen y evita que los valores que faltan cambien el rumbo de un análisis
pb.sustitucion_bfill(df, ['host_is_superhost', 'host_verifications', 'host_has_profile_pic', 'host_identity_verified'])
```
